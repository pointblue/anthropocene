# Finding the temporal autocorrelation order and evidence of slope
This notebook fits an ARIMA model for ADPE, EMPE, and WESE data looking for evidence of change

```r
# Load libraries
library(forecast); library(ggplot2);library(plyr)
```

```r
# Load the data
master<-read.csv("~/Workspace/Anthropocene/data/master_data.csv")
```

```r
# Here's how we fit an ARIMA model
# The plan is to fit one for each species and location to determine if there is a trend
head(master)
```

```r
amdl<-auto.arima(master$ADPE_CROZ,seasonal=FALSE)
summary(amdl)
```

```r
mdl2<-Arima(master$ADPE_CROZ,order=c(2,0,0),include.mean=F,include.drift=T)
summary(mdl2)
```

```r
checkresiduals(mdl2)
```

## Plotting the corrected trend

```r
adcroz<-master$ADPE_CROZ; names(adcroz)<-master$year
```

```r
# Uncorrected trend
ggtsdisplay(adcroz)
```

```r
# Corrected trend
pdf<-data.frame(year=master$year,uncorr=mmdl$x,corr=fitted(mmdl))
ggplot(pdf,aes(x=year,y=uncorr)) + geom_point() + geom_line(aes(y=corr),color="blue")
```

## Repeat the above for other colonies and other species...

```r
covmx<-as.matrix(master[,c("nmSST_aw_RSP","mnAirTemp","mnWind")])
mdl3<-Arima(master$ADPE_CROZ,order=c(2,0,0),include.mean=F,include.drift=F,xreg=covmx)
summary(mdl3)
```

```r
covmx<-as.matrix(master[,c("nmSST_aw_RSP","mnAirTemp")])
mdl4<-Arima(master$ADPE_CROZ,order=c(2,0,0),include.mean=F,include.drift=T,xreg=covmx)
summary(mdl4)
```
